```python
# FizzBuzz

# Bitte Zähle und beginne mit 1
# Wenn die Zahl die wir eingeben sollen durch 3 teilbar ist, dann sollen wir Fizz eingeben.
# Wenn die Zahl die wir eingeben sollen durch 5 teilbar ist, dann sollen wir Buzz eingeben.
# Wenn die Zahl die wir eingeben sollen durch 3 und 5 teilbar ist, dann sollen wir FizzBuzz eingeben.

print("""
HElloooooo!
Wilkommen bei FizzBuzz.
""")

while True:

    zahl = 1
    streak = 0

    while True:

        # Wir beten den Nutzer um seine Eingabe
        user_eingabe = input(f"Ich sag {zahl} du sagst: ").strip().lower()

        # Wir bestimmen die Korrekte Antwort
        if zahl % 3 == 0 and zahl % 5 == 0:
            korrekte_antwort = "fizzbuzz"
        elif zahl % 3 == 0:
            korrekte_antwort = "fizz"
        elif zahl % 5 == 0:
            korrekte_antwort = "buzz"
        else:
            korrekte_antwort = str(zahl)

        if user_eingabe == korrekte_antwort:
          streak += 1
          zahl += 1
          print(f"✅Korrekt! ------ Combo:{streak}⭐")
        else:
          print(f"☠️Falsch. Korrekt wäre {korrekte_antwort}☠️")
          break

    neustarten = input("Möchtest du nochmal Spielen? (Ja/Nein): ").lower()

    if neustarten != "ja":
      print("Danke fürs Spielen, Bye!")
      break
```
