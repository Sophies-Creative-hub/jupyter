## Python Dictionaries: Musterlösung zu der Gruppenaufgabe

```python
def serie_hinzufuegen(catalog, titel, genre, bewertung):
    # Neues Dictionary erstellen und unter dem Schlüssel 'titel' in catalog speichern
    # catalog[titel] = {}
    # catalog[titel]["genre"] = genre
    # catalog[titel]["bewertung"] = bewertung

    catalog[titel] = {
        "genre": genre,
        "bewertung": bewertung
    }
    print(f"Serie '{titel}' erfolgreich hinzugefügt.")

def serie_entfernen(catalog, titel):
    # Die Serie aus dem Katalog löschen, wenn die Serie nicht vorhanden ist, die Variable auf None setzen
    deleted_title = catalog.pop(titel, None)

    if deleted_title:
      print(f"Serie '{titel}' erfolgreich gelöscht. ")
    else:
      print(f"Serie '{titel}' ist nicht im Katalog.")

def katalog_anzeigen(catalog):
    # Alle Titel + Bewertung ausgeben
    if not catalog:
        print("Katalog ist leer.")
    else:
        print("Aktueller Serienkatalog:")
        for title, details in catalog.items():
            print(f"- {title} (Bewertung: {details['bewertung']})")

def serie_suchen(catalog, titel):
    # Serie anzeigen oder "Nicht vorhanden"
    if titel in catalog:
        details = catalog[titel]
        print(f"Gefunden: {titel} | Genre: {details['genre']} | Bewertung: {details['bewertung']}")
    else:
        print(f"Serie '{titel}' ist nicht vorhanden.")

def main():
    catalog = {}

    serie_hinzufuegen(catalog, "Stranger Things", "Mystery", 9.0)
    serie_hinzufuegen(catalog, "The Witcher", "Fantasy", 8.7)
    serie_hinzufuegen(catalog, "Dark", "Sci-Fi", 8.6)

    katalog_anzeigen(catalog)

    serie_suchen(catalog, "The Witcher")

    serie_entfernen(catalog, "Dark")
    katalog_anzeigen(catalog)

if __name__ == "__main__":
    main()
```
