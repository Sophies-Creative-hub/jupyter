# Dictionaries

Schlüssel-Wert-Paare

```python
serie = {
    "titel": "Stranger Things",
    "genre": "Mystery",
    "bewertung": 9.2
}

catalog = [
    {
        "titel": "Stranger Things",
        "genre": "Mystery",
        "bewertung": 9.2,
        "staffeln": 3
    },
    {
        "titel": "Vikings",
        "genre": "Action",
        "bewertung": 8.6
    },
    {
        "titel": "Game of Thrones",
        "genre": "Fantasy",
        "bewertung": 9.5,
        "staffeln": 8,
        "einzelne_titeln": [
            "Winter is coming",
            "The Kingsroad",
            "Lord Snow",
            "Cripples, Bastards, and Broken Things",
            "The Wolf and the Lion",
            "A Golden Crown",
            "You Win or You Die",
            "The Pointy End"
        ]
    }
]

print(serie)
print(catalog)
```

Dictionary von Dictionaries

```python
alle_serien = {
    "Stranger Things": {
        "genre": "Mystery",
        "bewertung": 9.2,
        "staffeln": 5
    },
    "Vikings": {
        "genre": "Action",
        "bewertung": 8.6
    },
    "Game of Thrones": {
        "genre": "Fantasy",
        "bewertung": 9.5,
        "staffeln": 8,
        "einzelne_titeln": [
            "Winter is coming",
            "The Kingsroad",
            "Lord Snow",
            "Cripples, Bastards, and Broken Things",
            "The Wolf and the Lion",
            "A Golden Crown",
            "You Win or You Die",
            "The Pointy End"
        ]
    }
}

print("Alle Serien:")
print(alle_serien)
print("Game of Thrones:")
print(alle_serien["Game of Thrones"])
print("Einzelne Episoden in Game of Thrones:")
print(alle_serien["Game of Thrones"]["einzelne_titeln"])
```

1. Einen Wert auslesen (Zugriff über Key)

```python
print(alle_serien["Stranger Things"])
```

```python
print(alle_serien["Stranger Things"]["bewertung"])
```

2. Wert überschreiben

Einen Wert mit einer Zuweisung überschreiben

```python
# Beretung von Stranger Things anpassen
alle_serien["Stranger Things"]["bewertung"] = 9.5

print(alle_serien["Stranger Things"])
print(alle_serien["Stranger Things"]["bewertung"])
```

```python
# Den Titel bei Game of Thrones anpassen
alle_serien["Game of Thrones"]["einzelne_titeln"][0] = "Winter is Here."

print(alle_serien["Game of Thrones"])
```

```python
# Noch einen Titel am Ende der Liste einsetzen
alle_serien["Game of Thrones"]["einzelne_titeln"].append("It was not the end.")

print(alle_serien["Game of Thrones"])
```

3. Ein neues Schlüssel-Wert Paar in das vorhandene Dictionary hinzufügen

```python
alle_serien["Stranger Things"]["jahr"] = 2020

print(alle_serien["Stranger Things"])

alle_serien["Stranger Things"]["cast"] = ["Schauspieler1", "Schauspieler2"]

print(alle_serien["Stranger Things"])
```

4. Einen Schlüssel entfernen


```python
del alle_serien["Vikings"]

print(alle_serien)
```

pop("Vikings") entfernt "Vikings" und gibt dessen Wert zurück. Wenn der Schluessel "Vikings" nicht existiert, kannst du optional einen zweiten Parameter (z.B. None) angeben, der dann zurückgegeben wird, damit kein Fehler entsteht.

```python
deleted_entry = alle_serien.pop("Vikings", "Wurde nicht gefunden.")

print(deleted_entry)

deleted_entry = alle_serien.pop("Vikings", "Wurde nicht gefunden.")

print(deleted_entry)
```

```python
del alle_serien["Vikings"]
```

5. Vorhandensein eines Keys prüfen



```python
if "Vikings" in alle_serien:
  print("Vikings ist vorhanden")
else:
  print("Vikings nicht vorhanden")

if "Stranger Things" in alle_serien:
  print("Stranger Things ist vorhanden")
else:
  print("Stranger Things nicht vorhanden")
```

6. Alle Schlüssel iterieren

```python
for key in alle_serien:
    print(key, "→", alle_serien[key])
```

7. Alle Werte iterieren

```python
for value in alle_serien.values():
    print(value)
```

8. Alle Schlüssel und Werte iterieren

```python
for key, value in alle_serien.items():
    print("Key: ", key)
    print("Value: ", value)
```
