# Die Schatzinsel

---

## **Anleitung zum "Schatzinsel-Spiel"**

### **Ziel des Spiels**

Das Ziel des Spiels ist es, den Schatz zu finden. Auf deinem Abenteuer wirst du Entscheidungen treffen müssen, die bestimmen, ob du den Schatz erreichst oder scheiterst. Jede Entscheidung führt zu unterschiedlichen Ereignissen und Enden. Deine Aufgabe ist es, die richtigen Entscheidungen zu treffen, um das Spiel zu gewinnen.

### **1. Einführung und Start**

Zu Beginn des Spiels wird dir eine Einführung angezeigt, die eine künstlerische Darstellung der Schatzinsel zeigt. Anschließend bekommst du eine kurze Einführung in das Spiel:

* **"Willkommen auf der Schatzinsel."**
* **"Deine Mission ist es, den Schatz zu finden."**

### **2. Die erste Entscheidung: Kreuzung**

Du befindest dich an einer Kreuzung und musst dich entscheiden, ob du nach **links** oder **rechts** gehen möchtest.

* Wenn du **"rechts"** wählst, wirst du von einem wilden Dingo angegriffen und das Spiel ist sofort vorbei.
* Wenn du **"links"** wählst, wirst du sicher weiterkommen und die nächste Etappe des Spiels erreichen.

**Tipp:** Wähle **"links"**, um voranzukommen.

### **3. Die zweite Entscheidung: Der See**

Nun stehst du vor einem See. Es gibt eine Insel in der Mitte des Sees, aber du kannst nicht einfach hinübergehen. Du hast zwei Optionen:

* Du kannst **"warten"** und auf ein Boot warten, das dich zur Insel bringt.

* Du kannst **"schwimmen"** und versuchen, selbst zum anderen Ufer zu gelangen.

* Wenn du **"schwimmen"** wählst, wirst du von den Sirenengesängen verführt und ertrinkst im See – das Spiel ist vorbei.

* Wenn du **"warten"** wählst, wirst du sicher zur Insel gebracht und kannst fortfahren.

**Tipp:** Wähle **"warten"**, um sicher die Insel zu erreichen.

### **4. Die dritte Entscheidung: Die Türen**

Du hast die Insel erreicht und findest ein Haus mit drei Türen: eine **rote**, eine **gelbe** und eine **blaue**. Jede Tür führt zu einem anderen Ergebnis:

* Wenn du die **blaue** Tür wählst, betrittst du einen Raum voller Welpen. Sie sind so niedlich, dass du die Kontrolle verlierst und das Spiel aufgrund der niedlichen Ablenkung verlierst.
* Wenn du die **gelbe** Tür wählst, findest du den **Schatz**! Du hast das Spiel gewonnen!
* Wenn du die **rote** Tür wählst, betrittst du einen Raum voller Türen, aber du kannst keinen Ausweg finden. Das Spiel ist vorbei.
* Wenn du eine Tür wählst, die nicht existiert (z. B. eine andere Farbe oder keine Auswahl), verlierst du auch das Spiel.

**Tipp:** Wähle die **"gelbe"** Tür, um den Schatz zu finden und das Spiel zu gewinnen!

### **Zusammenfassung der Entscheidungen:**

1. An der **Kreuzung** gehst du **links**.
2. Am **See** wartest du auf das **Boot**.
3. Bei den **Türen** wählst du die **gelbe** Tür, um den Schatz zu finden.

---

### **Fazit**

Das Spiel führt dich durch verschiedene Herausforderungen und Entscheidungen. Die richtigen Entscheidungen führen zum Sieg, während falsche Entscheidungen das Spiel vorzeitig beenden. Die Lösung ist es, in jeder Phase die richtige Wahl zu treffen und den Schatz am Ende zu finden.

#### **Schlüsselentscheidungen:**

* **Links gehen**
* **Warten auf das Boot**
* **Gelbe Tür wählen**

## Coding Hints

### **Schritt 1: Die Einführung des Spiels (zeige\_intro)**

Zu Beginn des Spiels möchtest du dem Spieler eine Einführung bieten, sowohl visuell als auch in Textform.

* **Was muss getan werden?**

  * Erstelle eine Funktion `zeige_intro()`, die eine ASCII-Grafik und eine kurze Willkommensnachricht ausgibt.
  * Diese Funktion wird direkt zu Beginn des Spiels aufgerufen, um den Spieler willkommen zu heißen und den Kontext zu erklären.

**Code-Beispiel:**

```python
def zeige_intro():
    """Zeigt die Einführungsgrafik und die Willkommensnachricht an."""
    print('''
    *******************************************************************************
    ... (hier kommt deine ASCII-Grafik) ...
    *******************************************************************************
    ''')
    print("Willkommen auf der Schatzinsel.")
    print("Deine Mission ist es, den Schatz zu finden.")
```

### **Schritt 2: Erste Entscheidung (waehle\_richtung)**

Der Spieler steht an einer Kreuzung und muss sich entscheiden, ob er **links** oder **rechts** gehen möchte.

* **Was muss getan werden?**

  * Erstelle eine Funktion `waehle_richtung()`, die den Spieler nach seiner Entscheidung fragt.
  * Wenn der Spieler **"rechts"** wählt, endet das Spiel sofort (Game Over).
  * Wenn der Spieler **"links"** wählt, darf er weitermachen.



### **Schritt 3: Zweite Entscheidung – Der See (ueberquere\_see)**

Der Spieler hat nun einen See erreicht und muss entscheiden, ob er **schwimmen** oder **warten** möchte.

* **Was muss getan werden?**

  * Erstelle eine Funktion `ueberquere_see()`, die den Spieler fragt, ob er schwimmen oder warten möchte.
  * Wenn der Spieler **"schwimmen"** wählt, endet das Spiel (Game Over), da er von Sirenengesängen gelockt wird.
  * Wenn der Spieler **"warten"** wählt, fährt das Spiel fort.


### **Schritt 4: Die Entscheidung mit den Türen (waehle\_tuer)**

Der Spieler steht nun vor einem Haus mit drei Türen und muss entscheiden, welche er öffnen möchte.

* **Was muss getan werden?**

  * Erstelle eine Funktion `waehle_tuer()`, die den Spieler fragt, welche Tür er wählen möchte: rot, blau oder gelb.
  * Jede Tür hat ein eigenes Ergebnis:

    * Blau: Der Raum voller Welpen lässt den Spieler nicht weiterkommen.
    * Gelb: Der Spieler findet den Schatz und gewinnt.
    * Rot: Der Spieler betritt einen Raum voller Türen, aber findet keinen Ausweg.


### **Schritt 5: Hauptspiel-Funktion (schatzinsel\_spiel)**

Nun müssen alle Funktionen zusammengeführt werden, um das vollständige Spiel zu gestalten. Du musst sicherstellen, dass jede Funktion nacheinander aufgerufen wird, je nachdem, welche Entscheidung der Spieler trifft.

* **Was muss getan werden?**

  * Erstelle eine Funktion `schatzinsel_spiel()`, die alle anderen Funktionen in der richtigen Reihenfolge aufruft:

    1. `zeige_intro()`
    2. `waehle_richtung()`
    3. `ueberquere_see()`
    4. `waehle_tuer()`



### **Schritt 6: Spiel starten**

Zum Schluss musst du das Spiel starten, indem du die Hauptspiel-Funktion aufrufst.

**Code-Beispiel:**

```python
# Spiel starten
schatzinsel_spiel()
```

### **Zusammenfassung:**

1. **zeige\_intro()** - Einführung und Willkommensnachricht anzeigen.
2. **waehle\_richtung()** - Der Spieler trifft die erste Entscheidung: links oder rechts.
3. **ueberquere\_see()** - Der Spieler entscheidet, ob er schwimmt oder auf ein Boot wartet.
4. **waehle\_tuer()** - Der Spieler trifft die Entscheidung, welche Tür er öffnen möchte.
5. **schatzinsel\_spiel()** - Die Hauptspiel-Funktion, die alle anderen Funktionen in der richtigen Reihenfolge aufruft.

Mit dieser Struktur kannst du das Spiel selbst schreiben und die Spiellogik Schritt für Schritt aufbauen!
