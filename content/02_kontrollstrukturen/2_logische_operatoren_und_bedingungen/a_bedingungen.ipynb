```python
# Boolean: Datentyp mit 2 Werte: True (1), False (0)

zahl = 10

ist_eingelogt = True # wenn der Benutzer eingellogt ist, dann hat die Variable den Wert True, wenn nicht - dann False
zahlung_im_verzug = False

positiv = zahl > 0 # True
negativ = zahl < 0 # False

print(ist_eingelogt)
print(zahlung_im_verzug)

print("\"Positiv\": ", positiv)
print('"Negativ": ', negativ)
```

```python
alter = int(input("Wie alt bist du? "))

# Vergleichsoperatoren
# alter >= 18 größer-gleich
# alter < 18 strikt kleiner
# alter == 18 gleich
# alter != 18 ungleich

ticketpreis = 15
ticketpreis_mit_rabatt = ticketpreis * 0.5

if alter >= 18:         # der 1. Fall: alter >= 18
  print("Du darfst den Film sehen.")
  print(f"Dein Ticketpreis ist {ticketpreis} Euro.")

else:                   # der 2. Fall: alter < 18
  print("Du darfst den Film mit deinen Eltern sehen.")
  print(f"Dein Ticketpreis ist {ticketpreis_mit_rabatt} Euro.")


print("Ende des Programms.")
```

```python
apfelsorte1 = ["süß", "weich"]
apfelsorte2 = ["sauer","weich"]
apfelsorte3 = ["süß", "hart"]
apfelsorte4 = ["sauer", "hart"]
```

```python
geschmack = input("Welchen Apfelgeschmack magst du? (gebe süß oder sauer ein): ")
konsistenz = input("Welche Konsistenz ist dir am liebsten? (gebe weich oder hart ein): ")

apfelsorte1_geschmack = "süß"
apfelsorte1_konsistenz = "weich"

apfelsorte2_geschmack = "sauer"
apfelsorte2_konsistenz = "weich"

apfelsorte3_geschmack = "süß"
apfelsorte3_konsistenz = "hart"

apfelsorte4_geschmack = "sauer"
apfelsorte4_konsistenz = "hart"


if geschmack == apfelsorte1_geschmack and konsistenz == apfelsorte1_konsistenz:
  print(f"Du magst Apfelsorte 1.")
elif geschmack == apfelsorte2_geschmack and konsistenz == apfelsorte2_konsistenz:
  print(f"Du magst Apfelsorte 2.")
elif geschmack == apfelsorte3_geschmack and konsistenz == apfelsorte3_konsistenz:
  print(f"Du magst Apfelsorte 3.")
elif geschmack == apfelsorte4_geschmack and konsistenz == apfelsorte4_konsistenz:
  print(f"Du magst Apfelsorte 4.")
else:
  print("Es gibt keine passende Apfelsorte.")



```
