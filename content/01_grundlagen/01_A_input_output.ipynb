
````python
# TODO: Geschwindigkeit eingeben, Fahrtzeit und Pausen berechnen
geschwindigkeit = float(input("Wie schnell f√§hrst du (km/h)? "))
strecke = float(input("Wie lang ist die Strecke in km? "))

fahrzeit = strecke / geschwindigkeit

pausen = int((fahrzeit - 0.01 ) // 2)

gesamtzeit = fahrzeit + pausen * 0.25

print(f"Du brauchst {fahrzeit} Stunden Fahrzeit.")

print(f"Mit {pausen} Pause(n) √† 15 Minuten dauert die gesamte Reise ca. {gesamtzeit} Stunden.")
````
