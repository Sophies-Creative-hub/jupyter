```python
def monat_name(monatszahl):
    monate = [
        "Januar", "Februar", "März", "April", "Mai", "Juni",
        "Juli", "August", "September", "Oktober", "November", "Dezember"
    ]
    if 1 <= monatszahl <= 12:
        return monate[monatszahl - 1]
    else:
        return "Ungültiger Monat"

# Test

print(monat_name(3)) # Sollte "März" ausgeben
print(monat_name(12)) # Sollte "Dezember" ausgeben
print(monat_name(13)) # Sollte "Ungültiger Monat" ausgeben
```
