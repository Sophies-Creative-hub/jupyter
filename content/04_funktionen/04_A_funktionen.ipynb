```python
# Level 3.1: Schreiben Sie eine Funktion, die die Summe einer Liste berechnet
def summe_liste(zahl):
  return sum(zahl)

print(summe_liste([1, 2, 3, 4])) # Sollte 10 ausgeben
print(summe_liste([5, 10, 15])) # Sollte 30 ausgeben
```

```python
# Level 3.2: Schreiben Sie eine Funktion, die eine Liste mit Zahlen von 1 bis n erstellt
def liste_bis_n(zahl):
  return list(range(1, zahl, +1))

print(liste_bis_n(5)) # Sollte [1, 2, 3, 4, 5] ausgeben
print(liste_bis_n(3)) # Sollte [1, 2, 3] ausgeben
```

```python
# Level 3.3: Schreiben Sie eine Funktion, die eine neue Liste mit den quadrierten Werten einer Liste erstellt
def quadriere_liste(liste):
  quadrate = []
  for zahl in liste:
    quadrate.append(zahl ** 2)
  return quadrate

print(quadriere_liste([1, 2, 3])) # Sollte [1, 4, 9] ausgeben
print(quadriere_liste([4, 5])) # Sollte [16, 25] ausgeben
```

```python
# Level 3.4: Schreiben Sie eine Funktion, die das Produkt aller Zahlen in einer Liste berechnet
def produkt_liste(liste):
  produkt = 1
  for zahl in liste:
    produkt *= zahl
  return produkt


print(produkt_liste([1, 2, 3, 4])) # Sollte 24 ausgeben
print(produkt_liste([5, 2])) # Sollte 10 ausgeben
```

```python
# Level 3.5: Schreiben Sie eine Funktion, die das Minimum einer Liste zurückgibt
def minimum_liste(liste):
  return min(liste)

print(minimum_liste([5, 3, 8, 1, 9])) # Sollte 1 ausgeben
print(minimum_liste([10, 20, 5, 30])) # Sollte 5 ausgeben
```

```python
# Level 3.6: Schreiben Sie eine Funktion, die zählt, wie viele ungerade Zahlen in einer Liste sind
def zaehle_ungerade(liste):
  counter = 0
  for zahl in liste:
    if zahl % 2 != 0:
        counter += 1
  return counter


print(zaehle_ungerade([1, 2, 3, 4, 5])) # Sollte 3 ausgeben
print(zaehle_ungerade([2, 4, 6])) # Sollte 0 ausgeben
```

```python
# Level 3.7: Schreiben Sie eine Funktion, die prüft, ob alle Elemente einer Liste positiv sind
def alle_positiv(liste):
  counter = 0
  for zahl in liste:
    if zahl > 0:
      counter += 1
  if counter == len(liste):
    return True
  else:
    return False

  return all(value > 0 for value in liste)

print(alle_positiv([1, 2, 3])) # Sollte True ausgeben
print(alle_positiv([1, 0, 3])) # Sollte False ausgeben
print(alle_positiv([1, -2, 3])) # Sollte False ausgeben
```

```python
# Level 3.8: Schreiben Sie eine Funktion, die eine Liste mit geraden Zahlen von 2 bis 2n erstellt
def gerade_bis_2n(n):
  return list(range(2, 2 * n +1, 2))

print(gerade_bis_2n(5)) # Sollte [2, 4, 6, 8, 10] ausgeben
print(gerade_bis_2n(3)) # Sollte [2, 4, 6] ausgeben
```

```python
# Level 3.9: Schreiben Sie eine Funktion, die eine Liste nach Vorkommen eines Elements durchsucht und dessen Position ausgibt.
def finde_index(liste, zahl):

  if zahl in liste:
    return liste.index(zahl)
  else:
    return -1

print(finde_index([10, 20, 30, 40], 30)) # Sollte 2 ausgeben
print(finde_index([10, 20, 30, 40], 50)) # Sollte -1 ausgeben
```
