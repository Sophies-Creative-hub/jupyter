## Wetterbericht

**Lernziel:** Eine Funktion mit einem Parameter und bedingten Anweisungen

**Aufgabe:**

1. Erstelle eine Funktion `zeige_wetter(wetter_typ)`, die je nach Wetter unterschiedliche ASCII-Art anzeigt
2. Frage nach dem Wetter (sonnig, regnerisch, bewölkt) und rufe die Funktion auf

**Pseudocode:**

```
Definiere Funktion zeige_wetter(wetter_typ):
    Wenn wetter_typ ist "sonnig":
        Zeige Sonnen-ASCII-Art
    Sonst wenn wetter_typ ist "regnerisch":
        Zeige Regen-ASCII-Art
    Sonst wenn wetter_typ ist "bewölkt":
        Zeige Wolken-ASCII-Art
    Sonst:
        Zeige unbekanntes Wetter

# Hauptprogramm
Frage nach dem Wetter (sonnig, regnerisch, bewölkt)
Rufe zeige_wetter mit dem eingegebenen Wetter auf
```

**ASCII-Art für "sonnig":**

```
    \   /
     .-.
  ― (   ) ―
     `-'
    /   \
```

**ASCII-Art für "regnerisch":**

```
     .-.
    (   ).
   (___(__)
    ' ' ' '
   ' ' ' '
```

**ASCII-Art für "bewölkt":**

```
   .--.
 .(    ).
(___.__)__)
```
