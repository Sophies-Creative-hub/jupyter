#### Erweiterung mit Begrenzungen und while-Loop
*Optionale Erweiterung:*

**Hunger und Glück mit Grenzen:**

Der Wert des "Hunger"-Attributs darf niemals unter 0 fallen. Achte darauf, dass das Haustier nur dann gefüttert wird, wenn der Hungerwert größer als 0 ist.

Der Wert des "Glück"-Attributs darf maximal 10 betragen. Erhöhe das Glück nur bis zu diesem Wert.

**Eingabe-Schleife:**

Benutze einen while-Loop, damit der Benutzer das Haustier kontinuierlich füttern oder mit ihm spielen kann, bis er das Programm beendet.

**Pseudocode**

```
Definiere Funktion fuettere_haustier(haustier):
    Wenn hunger > 0, dann hunger verringern (aber nicht unter 0)
    Bestätigungsnachricht ausgeben
    Haustier zurückgeben

Definiere Funktion spiele_mit_haustier(haustier):
    Wenn glueck < 10, dann glueck erhöhen (aber nicht über 10)
    Bestätigungsnachricht ausgeben
    Haustier zurückgeben

# Hauptprogramm mit Schleife
Nach Name und Tierart fragen
Haustier erstellen
Haustier anzeigen

Solange der Benutzer das Programm nicht beenden möchte:
    Nach Aktion fragen (füttern oder spielen oder beenden)
    Wenn füttern:
        fuettere_haustier(haustier)
    Wenn spielen:
        spiele_mit_haustier(haustier)
    Wenn beenden:
        Schleife beenden
    Aktuelles Haustier anzeigen
```

image.png

#### Erklärung der Erweiterungen

**Grenzen für Hunger und Glück:**

Die Funktion `fuettere_haustier()` sorgt dafür, dass der Hunger des Haustiers nur verringert wird, wenn der Hunger größer als 0 ist. Ansonsten wird eine Nachricht angezeigt, dass das Haustier nicht hungrig ist.

In der Funktion `spiele_mit_haustier()` wird das Glück des Haustiers nur erhöht, wenn es noch nicht den maximalen Wert von 10 erreicht hat. Falls das Haustier bereits glücklich genug ist, wird dies mit einer Nachricht bestätigt.

**while-Loop:**

Der while-Loop sorgt dafür, dass das Programm immer weiterläuft, bis der Benutzer explizit das Programm beendet (durch Eingabe von "beenden").

Der Benutzer kann jederzeit entscheiden, ob er das Haustier füttern, mit ihm spielen oder das Programm beenden möchte.

## Zahlenraten-Spiel

**Lernziel:** Funktionen mit Parametern und Rückgabewerten, zufällige Zahlen, Schleifen

**Aufgabe:**

1. Schreibe eine Funktion `generiere_zahl(max_wert)`, die eine Zufallszahl zurückgibt
2. Erstelle eine Funktion `pruefe_versuch(geheimzahl, versuch)`, die den Versuch bewertet
3. Füge eine Funktion `zeige_ergebnis(ergebnis, versuche)` hinzu, die das Ergebnis anzeigt
4. Schreibe eine Hauptfunktion `spiele_zahlenraten()`, die das Spiel steuert

**Pseudocode:**

```
Definiere Funktion generiere_zahl(max_wert=10):
    Erzeuge eine Zufallszahl zwischen 1 und max_wert
    Gib die Zahl zurück

Definiere Funktion pruefe_versuch(geheimzahl, versuch):
    Wenn versuch < geheimzahl, gib "zu niedrig" zurück
    Wenn versuch > geheimzahl, gib "zu hoch" zurück
    Sonst gib "richtig" zurück

Definiere Funktion zeige_ergebnis(ergebnis, versuche):
    Wenn ergebnis == "richtig", zeige Gratulations-ASCII-Art
    Sonst zeige einen Hinweis mit Pfeil

Definiere Funktion spiele_zahlenraten():
    Zeige Spielanleitung
    Generiere eine Geheimzahl
    Initialisiere Versuche und erraten-Status

    Solange nicht erraten und Versuche < 5:
        Frage nach einem Tipp
        Erhöhe Versuche
        Prüfe den Versuch
        Zeige das Ergebnis
        Wenn richtig geraten, setze erraten auf wahr

    Wenn nicht erraten, zeige Game-Over und die Geheimzahl

# Hauptprogramm
Rufe spiele_zahlenraten auf
```

**ASCII-Art für deine Lösung:**

```
+-----------------------+
| ZAHLENRATEN           |
| Errate meine Zahl     |
| zwischen 1 und 10!    |
+-----------------------+
```

```
⭐⭐⭐⭐⭐⭐⭐⭐
⭐  GRATULATION!  ⭐
⭐  {versuche} Versuche!   ⭐
⭐⭐⭐⭐⭐⭐⭐⭐
```

```
+-----------------------+
| GAME OVER!            |
| Die Zahl war: {geheimzahl}     |
+-----------------------+
```
