## Was sind Funktionen?

Funktionen sind wiederverwendbare Codeblöcke, die eine bestimmte Aufgabe erfüllen. Sie helfen uns, Code zu organisieren, zu strukturieren und Wiederholungen zu vermeiden. Eine Funktion ist wie ein kleines Programm innerhalb unseres Hauptprogramms.

Vorteile von Funktionen:
- **Wiederverwendbarkeit**: Schreibe den Code einmal, verwende ihn mehrfach
- **Übersichtlichkeit**: Strukturiere deinen Code in logische Einheiten
- **Wartbarkeit**: Ändere die Funktionalität an einer zentralen Stelle
