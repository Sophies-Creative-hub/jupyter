## Rückgabewerte

### Konzept
Funktionen können nicht nur Aktionen ausführen, sondern auch Werte zurückgeben, die dann in anderen Teilen des Programms verwendet werden können.

```python
def add_numbers(a, b):
    return a + b

# Die Funktion gibt einen Wert zurück, den wir in einer Variable speichern können
result = add_numbers(5, 3)
print(result)  # Ausgabe: 8

# Oder wir können den Rückgabewert direkt verwenden
print(add_numbers(10, 20))  # Ausgabe: 30
```

Erklärung:
- Das Schlüsselwort `return` gibt einen Wert zurück
- Nach `return` wird die Funktion sofort beendet
- Der zurückgegebene Wert kann gespeichert oder direkt verwendet werden

## Standardparameter

### Konzept
Standardparameter erlauben es uns, Funktionen mit vordefinierten Werten für Parameter zu erstellen. Diese werden verwendet, wenn beim Funktionsaufruf kein expliziter Wert übergeben wird.

```python
def power(base, exponent=2):
    return base ** exponent

# Verwendung des Standardwerts für den zweiten Parameter
print(power(3))  # Berechnet 3², Ausgabe: 9

# Überschreiben des Standardwerts
print(power(3, 3))  # Berechnet 3³, Ausgabe: 27
```

Erklärung:
- Der Parameter `exponent` hat den Standardwert `2`
- Wenn wir die Funktion mit nur einem Argument aufrufen, wird der Standardwert verwendet
- Wir können den Standardwert jederzeit überschreiben, indem wir einen expliziten Wert angeben
