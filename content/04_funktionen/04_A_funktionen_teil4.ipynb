## Einfache Funktion ohne Parameter

### Konzept
Die einfachste Form einer Funktion hat keinen Parameter und führt einfach eine Aktion aus.

```python
def greet():
    print("Hallo, Welt!")

# Funktionsaufruf
greet()
```

Erklärung:
- `def` ist das Schlüsselwort, um eine Funktion zu definieren
- `greet` ist der Name der Funktion
- Die leeren Klammern `()` zeigen, dass die Funktion keine Parameter erwartet
- Der Code innerhalb der Funktion ist eingerückt
- Die Funktion wird durch Angabe ihres Namens gefolgt von `()` aufgerufen

## Funktionen mit Parametern

### Konzept
Funktionen werden noch nützlicher, wenn sie Daten als Eingabe erhalten können. Diese Eingaben nennt man Parameter oder Argumente.

```python
def greet(name):
    print(f"Hallo, {name}!")

# Funktionsaufruf mit einem Argument
greet("Maria")
greet("Thomas")
```

Erklärung:
- Der Parameter `name` ist wie eine Variable, die nur innerhalb der Funktion existiert
- Der Wert des Parameters wird beim Funktionsaufruf übergeben
- Mit derselben Funktion können wir verschiedene Personen begrüßen
