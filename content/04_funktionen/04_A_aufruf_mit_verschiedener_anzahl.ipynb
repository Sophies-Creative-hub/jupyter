## Variable Argumentlisten

### Konzept
Manchmal möchten wir eine Funktion mit einer unbestimmten Anzahl von Argumenten aufrufen. Python ermöglicht dies mit der `*args`-Syntax.

```python
def average(*numbers):
    return sum(numbers) / len(numbers)

# Aufruf mit verschiedener Anzahl von Argumenten
print(average(10, 20, 30))  # Ausgabe: 20.0
print(average(1, 2, 3, 4, 5))  # Ausgabe: 3.0
```

Erklärung:
- `*numbers` sammelt alle übergebenen Argumente in einem Tupel
- Die Funktion kann mit beliebig vielen Argumenten aufgerufen werden
- Innerhalb der Funktion wird `numbers` als Tupel behandelt

## Bedingte Anweisungen in Funktionen

### Konzept
Funktionen können Entscheidungen treffen und unterschiedliche Aktionen basierend auf Bedingungen ausführen.

```python
def check_even_odd(number):
    if number % 2 == 0:
        return "Gerade"
    else:
        return "Ungerade"

print(check_even_odd(7))  # Ausgabe: Ungerade
print(check_even_odd(10))  # Ausgabe: Gerade
```

Erklärung:
- Die Funktion verwendet eine `if-else`-Anweisung, um zu prüfen, ob die Zahl gerade oder ungerade ist
- Der Modulo-Operator `%` gibt den Rest einer Division zurück
- Abhängig vom Ergebnis der Bedingung wird ein unterschiedlicher Wert zurückgegeben
