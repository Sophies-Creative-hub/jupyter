# Drill 1.4: Schreibe eine Funktion, die prüft, ob eine Zahl positiv ist


```python
def ist_positiv(zahl):
  if zahl > 0:
    return True
  elif zahl < 0:
    return False
  else:
    return False


# Test
print(ist_positiv(5))      # Sollte True ausgeben
print(ist_positiv(-3))     # Sollte False ausgeben
print(ist_positiv(0))      # Sollte False ausgeben
```
