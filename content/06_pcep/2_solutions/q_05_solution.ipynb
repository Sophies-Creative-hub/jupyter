Frage 50

Welche Datenstruktur ist am besten geeignet, um die Häufigkeit des Vorkommens von Wörtern in einem Text zu speichern?

- [ ] Liste
- [ ] Tupel
- [x] Dictionary
- [ ] Set

## Begründung:
- Für Häufigkeitszählungen können wir am Besten Schlüssel-Wert-Paare nutzen
- Schlüssel: Wort und Wert: Häufigkeit
- ```python
  dict = {
    wort1: 4,
    wort2: 5,
    ...
  }
  ```

````python
text = "Der Lehrer sagte den Schülern, dass sie den Text, den er den Schülern gestern gegeben hatte, lesen sollten, denn der Text war wichtig für die Prüfung, die Prüfung, die nächste Woche stattfinden wird, wobei jeder Schüler, der den Text nicht gelesen hat, Probleme haben könnte, denn der Lehrer hatte mehrmals betont, wie wichtig dieser Text für das Verständnis des Themas ist, das Thema, das sie seit Wochen bearbeiten."
words = text.split()
count = {}
for w in words:
  count[w] = count.get(w,0) + 1
print(count)
````
