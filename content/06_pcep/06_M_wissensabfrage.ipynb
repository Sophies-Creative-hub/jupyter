**Frage:** Was macht diese Funktion?
```python

```

- a) Gibt immer 2 zurück
- b) Gibt `x` zurück

z.B. für x = 3

## 5.2 Verschachtelte Funktion
Was gibt die folgende verschachtelte Funktion zurück?
```python

def outer_function(x):
  def inner_function(y):
    return x * y + 2
  return inner_function(x + 1)
result = outer_function(3) # Was ist hier der Inhalt von result?
````
- 14
- 12
- 8
- 11

`````python
def outer_function(x):
  def inner_function(y):
    return x * y + 2 # x * (x + 1) + 2 bzw. 3 * (3+1) + 2
  return inner_function(x + 1)

# Erster Schritt: outer_function mit 3 aufrufen, d.h. x = 3 einsetzen, d.h. inner function ist dann 3 * y + 2
# Zweiter Schritt: setze für y in die innere Funktion (x+1, bzw. 3+1) ein.
result = outer_function(3)
print(f"Das Ergebnis des Funktionsaufrufs lautet {result}")
`````
