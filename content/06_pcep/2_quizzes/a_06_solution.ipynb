Was macht das `nonlocal`-Schlüsselwort in Python?

# Frage 26
Was ist die grundlegende Struktur für die Ausnahmebehandlung in Python?
- [ ] `catch-try-finally`
- [x] `try-except-else-finally`
- [ ] `try-catch-finally`
- [ ] `attempt-except-otherwise`

````python
try:
  print("Kein Fehler") # Dieser Blocker wird ausprobiert, wenn kein Fehler passiert, läuft der einfach weiter
except:
  print("Fehler") # Dieser Block wird nur ausgeführt wenn im try-Block ein Fehler auftritt. In diesem Fall fangen wir alle Fehler ab, da kein spezifischer Fehler angegeben ist
else:
  print("Alles gut") # Dieser Block wird nur ausgeführt wenn kein Fehler im try-Block passiert.
finally:
  print("Immer ausgeführt") # Dieser Block wird immer ausgeführt, egal ob es einen Fehler gab oder nicht. Das ist ganz gut zum Aufräumen von Aktionen, die immer passieren sollen
````

````python
while True:
  try:
    a = int(input("Bitte gib eine Zahl ein: "))
    b = int(input("Bitte gib eine Zahl ein: "))
    print(f"Das Ergebnis der Division ist {a/b}")
  except ZeroDivisionError:
    print("Teile nicht durch 0") # -> finally wird direkt hier hinter ausgeführt
    break # hiermit springen wir aus Schleife raus
  except ValueError:
    print("Bitte gib eine gültige Zahl ein")
  finally:
    print("das ist finally")
````

# Frage 45
Was ist die Ausgabe des folgenden Codes?
```python
def outer():
    x = 1
    def inner():
        nonlocal x
        x = 2
        print("Inner:", x)
    inner()
    print("Outer:", x)

outer()
```
- [ ] Inner: 1, Outer: 1
- [ ] Inner: 2, Outer: 1
- [ ] Inner: 2, Outer: 2
- [ ] Ein Fehler wird ausgelöst

````python
def outer():
    x = 1
    def inner():
        nonlocal x # nonlocal bezieht sich auf das x aus der umschließenden Funktion
        x = 2
        print("Inner:", x)
    inner()
    print("Outer:", x)

outer()

## eine Variable wie z.B. x ist in einer Funktion immer local, das bedeutet ihr Scope/Reichweite bezieht sich nur auf die Funktion nur innerhalb unserer Funktion
## wenn ich ihr aber ein nonlocal in einer inneren Funktion zuweise, dann heißt das, dass ich mich auf eine Variable aus dem Kontext der äußeren Funktion beziehe
## ebenso bei global: wenn ich keine inneren oder äußeren Funktion habe, sondern ganz normal eine Funktion, macht es dasselbe wie nonlocal, bezieht sich also auf das Objekt außerhalb meiner Funktionsreichweite
````

````python
x = 5 # das nennen wir auch "globale Variable"

def test():
  global x
  # nonlocal x # das funktioniert nicht, weil wir keine äußere Funktion haben
  print(x)

test()

## global x und nonlocal x haben dieselbe Idee, nonlocal ist nur der Spezialfall für innere Funktion
````

# Frage 50
Was ist die Ausgabe des folgenden Codes?
```python
def counter():
    count = 0
    def increment():
        nonlocal count
        count += 1
        return count
    return increment

my_counter = counter()
print(my_counter())
print(my_counter())
```
- [ ] 0, 0
- [ ] 0, 1
- [x] 1, 2
- [ ] Ein Fehler wird ausgelöst

## Begründung:


````python
def counter():
    count = 0
    def increment():
        nonlocal count
        count += 1
        return count
    return increment

my_counter = counter() # Instanziierung der Funktion counter() durch die Umbenennung
my_second_counter = counter()
print("Zähler vom ersten Counter",my_counter()) # Erster Aufruf der counter-Funktion, setzt den counter auf 1
# ab jetzt ist count = 1
#print(count)
print("Zähler vom ersten Counter",my_counter()) # Zweiter Aufruf inkrementiert den counter erneut um 1, bedeutet counter = 2 jetzt
print("Zähler vom zweiten Counter",my_second_counter())
````
